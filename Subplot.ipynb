```python
from matplotlib import pyplot as plt

# Set the figure size
plt.rcParams["figure.figsize"] = [7.00, 3.50]
plt.rcParams["figure.autolayout"] = True

# Create a new figure using plt.figure
fig1 = plt.figure("Figure 1 - plt.figure")

# Create a figure and set of subplots using plt.subplots
fig2, ax = plt.subplots(nrows=2,ncols=3,sharex=True,sharey=True)

ax[0][0].plot(['A','B','C','D'],[1,2,3,20])
ax[0][1].bar(['A','B','C','D'],[1,2,3,20])
ax[0][2].scatter(['A','B','C','D'],[1,2,3,20])
ax[1][0].barh(['A','B','C','D'],[1,2,3,20],color='green')
ax[1][0].barh(['A','B','C','D'],[10,12,30,20])
# Display the plot
plt.show()
```
