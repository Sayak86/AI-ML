```python
dfnew = pd.read_csv('DataSets/updated_EDA_census.csv')
x = dfnew[
    (dfnew['Age group'] == '20-24')  & 
    (dfnew['Area Name'] == 'INDIA') & 
    (dfnew['Total/Rural/Urban'] == 'Total')
    ]

(x['Total Female Illiterate'] / x['Total Female Population']) *100
```

```python
dfnew['Literacy Rate'] = dfnew[
    (dfnew['Area Name'] == 'INDIA') & 
    (dfnew['Total/Rural/Urban'] == 'Total')
    ].groupby(by='Age group')[['Total Literate','Total Population']].apply(lambda a:a['Total Literate']/a['Total Population']*100)
```

```python

dfnew['Literacy rate']= (dfnew['Total Literate'] /dfnew['Total Population']) * 100
dfnew[['Age group','Total Literate','Total Population','Literacy rate','Area Name']]
```

```python
pd.pivot_table(dfnew,index=['Age group'],values=['Literacy rate'])
```

```python
female_literacy = dfnew[ (dfnew['Area Name'] !='INDIA') & (dfnew['Total/Rural/Urban'] == 'Total') & 
                        (dfnew['Age group'] == 'All ages')]
female_literacy[['Literacy rate','Area Name']].sort_values(by='Literacy rate',ascending=False)
```
